```julia
using Flux
```

```julia
A = Flux.Data.Iris.features()
```

```julia
X = convert(Array{Float64, 2}, A')
```

```julia
using Statistics
```

```julia
mean(X, dims=1)
```

```julia
feature_order = [1, 3, 2, 4]
```

```julia
X[:, feature_order]
```

```julia
using Plots
```

```julia
plot(X[:, 1], X[:, 2], st=:scatter)
```

```julia
y = Flux.Data.Iris.labels()
```

```julia
labels = unique(y)
```

```julia
begin
    setosa = (y .== labels[1])
    versicolor = (y .== labels[2])
    virginica = (y .== labels[3])
end
```

```julia
plot(X[setosa, 1], X[setosa, 2], st=:scatter, legend=false, m=:+, c=:black)
```

```julia
plot!(X[versicolor, 1], X[versicolor, 2], st=:scatter, legend=false, m=:*, c=:red)
```

```julia
plot!(X[virginica, 1], X[virginica, 2], st=:scatter, legend=false, m=:o, c=:yellow)
```

```julia
μ = mean(X, dims=1)
```

```julia
M = repeat(μ, 150, 1)
```

```julia
X̄ = X - M
```

```julia
mean(X̄, dims=1)
```

```julia
B = X̄[:, feature_order]
```

```julia
begin
    plot(B[setosa, 1], B[setosa, 2], st=:scatter, legend=false, m=:+, c=:black, xlabel="Sepal Lenght", ylabel="Petal Lenght")
    plot!(B[versicolor, 1], B[versicolor, 2], st=:scatter, legend=false, m=:*, c=:red)
    plot!(B[virginica, 1], B[virginica, 2], st=:scatter, legend=false, m=:o, c=:yellow)
end
```

```julia
Σ = cov(B)
```

```julia
B'*B/(length(y) - 1)
```

```julia
using LinearAlgebra
```

```julia
value_propre, vector_propre = eigen(Σ)
```

```julia
value_propre
```

```julia
vector_propre
```

```julia
order = sortperm(value_propre, rev=true)
```

```julia
θ = vector_propre[:, order[1:2]]
```

```julia
Z = B * θ
```

```julia
begin
    plot(Z[setosa, 1], Z[setosa, 2], st=:scatter, legend=false, m=:+, c=:black, xlabel="1st component", ylabel="2nd component")
    plot!(Z[versicolor, 1], Z[versicolor, 2], st=:scatter, legend=false, m=:*, c=:red)
    plot!(Z[virginica, 1], Z[virginica, 2], st=:scatter, legend=false, m=:o, c=:yellow)
end
```
