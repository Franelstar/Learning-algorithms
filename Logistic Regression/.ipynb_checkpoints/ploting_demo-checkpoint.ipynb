```julia
using Plots
#using PyPlot
```

```julia
θ = -3.0:0.1:3.0
```

```julia
J(z) = 2*z*z - 3*z + 1
```

```julia
y = J.(θ)
```

```julia
plot(θ, y)
```

```julia
θ_best = 3/4
```

```julia
y_best = J.(θ_best)
```

```julia
plot!([θ_best], [y_best], st=:scatter, color=:red)
```

```julia
θ_2 = -3:0.1:3
```

```julia
J(a, b) = 2*a*a - 3*a*b + b*b - 4
```
